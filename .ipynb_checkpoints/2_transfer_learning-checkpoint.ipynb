```python
import tensorflow as tf
import keras
from keras.models import Sequential, Model 
from keras.layers.core import Flatten, Dense, Dropout
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.convolutional import ZeroPadding2D
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras.optimizers import SGD
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
import data
import models
from validation import recall_m, precision_m, f1_m, auc
```

```python
data_marm,label_marm,data_marm_tl,label_marm_tl=data.marmousi()
```

```python
print('marmousi transfer learning input size =',data_marm_tl.shape)
print('marmousi transfer learning label size =',label_marm_tl.shape)

print('marmousi entire input size =',data_marm.shape)
print('marmousi entire label size =',label_marm.shape)
```

```python
import tensorflow as tf
from keras.models import load_model
model=load_model('CNN.hdf5', custom_objects={'tf': tf})

model.summary()
```

```python
prediction = model.predict(data_marm)
prediction_classes=np.argmax(prediction,axis=1)
predicted_vel_base_model=np.zeros([385,45])
for i in range(385):
    predicted_vel_base_model[i,:]=prediction_classes[i*45:i*45+45]
```

```python
n=0
x=25
fig = plt.figure(figsize=[15,15])
ax1 = fig.add_subplot(3,1,1)
ax1.imshow(predicted_vel_base_model.T,aspect='auto',vmin=n,vmax=x)
ax1.title.set_text('prediction without transfer learning')
```

```python
for layer in model.layers[:4]:
    layer.trainable=False
```

```python
for i,layer in enumerate(model.layers):
    print(i,layer.name, layer.trainable)
```

```python
def learning_rate_decay(epoch):
    initial_lr = 1e-4
    max_epoch = 20

    if epoch < max_epoch:
        lr = initial_lr
    else:
        idecay = (epoch // max_epoch)
        lr = initial_lr/(idecay * 2)
    return lr
lr_sched = LearningRateScheduler(learning_rate_decay)
model_checkpoint = ModelCheckpoint('CNN_transfer.hdf5', monitor='loss', verbose=0, save_best_only=True)
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['acc'])
```

```python
history=model.fit(data_marm_tl, label_marm_tl, 
                  validation_data=(data_marm, label_marm), 
                  nb_epoch=100, 
                  batch_size=315, 
                  callbacks=[model_checkpoint,lr_sched ],
                  verbose=1)
```

```python
prediction = model.predict(data_marm)
```

```python
prediction_classes=np.argmax(prediction,axis=1)
true_classes=np.argmax(label_marm,axis=1)
```

```python
predicted_vel=np.zeros([385,45])
true_vel=np.zeros([385,45])
for i in range(385):
    predicted_vel[i,:]=prediction_classes[i*45:i*45+45]
    true_vel[i,:]=true_classes[i*45:i*45+45]
```

```python
n=0
x=25
fig = plt.figure(figsize=[15,30])
ax1 = fig.add_subplot(4,1,1)
ax1.imshow(predicted_vel_base_model.T,aspect='auto',vmin=n,vmax=x)
ax1.title.set_text('prediction without transfer learning')
ax2 = fig.add_subplot(4,1,2)
ax2.imshow(predicted_vel.T,aspect='auto',vmin=n,vmax=x)
ax2.title.set_text('prediction with transfer learning using 10% of target data')
ax3 = fig.add_subplot(4,1,3)
ax3.imshow(true_vel.T,aspect='auto',vmin=n,vmax=x)
ax3.title.set_text('true')
ax4 = fig.add_subplot(4,1,4)
ax4.imshow(np.abs(predicted_vel.T-true_vel.T),aspect='auto',vmin=n,vmax=x)
ax4.title.set_text('abs(prediction-true)')
```
